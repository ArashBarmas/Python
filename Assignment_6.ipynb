```python
# RGB
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
```

```python
img_white = mpimg.imread('a.jpg')
img_middle = mpimg.imread('b.jpg')
plt.imshow(img_middle)
img_middle.shape
```

```python
# Question 1
img_w = img_white[:,:,0:3].copy()
img_w[(297 - 320/2):(297 - 320/2+320),(250 - 120): (250 -120+240)] = img_middle
plt.imsave("c.jpg",img_w)
plt.imshow(img_w)
```

```python
# Question 3
img_g = mpimg.imread("g.jpg")
plt.imshow(img_g)
```

```python
img_minion = img_g[:,:,0:3].copy()
#img_minion[img_minion[:,:,1]==255,1] = 0
```

```python
mask = (img_minion[:,:,1]>=225) & (img_minion[:,:,2]<50) & (img_minion[:,:,0]<50)
img_minion[mask,1] = 0
plt.imshow(img_minion)
img_minion.shape
```

```python
img_h = mpimg.imread("h.jpg")
plt.imshow(img_h)
img_h.shape
img_final = img_h[:,:,0:3].copy()
```

```python
# not mask is the minion 
img_final[1137:1137 +530,(2500/2-330):(2500/2-330+660),0:3][np.logical_not(mask)] = img_minion[np.logical_not(mask)]
plt.imshow(img_final)
```

```python
plt.imsave('minion.jpg',img_final)
```
